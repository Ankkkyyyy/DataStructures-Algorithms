




If there are odd number of elements, return the middle element,   
if there are even elements return the one which is further from the head node.

#### 2 pointer or floyd cycle algo

```python
class Node:
    def __init__(self,data):
        self.data = data 
        self.next = None   
```

```python
class LinkedList:
    def __init__(self):
        self.head = None 
    
    def print(self):
        temp = self.head 
        while temp is not None:
            print(temp.data)
            temp = temp.next

    def findMiddle(self):
        slow = self.head 
        fast = self.head 
        while(fast!=None and fast.next!=None):
            slow = slow.next 
            fast = fast.next.next 
        return slow.data 
```

```python
ll = LinkedList()
first = Node(10)
second = Node(20)
third = Node(30)
# fourth = Node(99)
ll.head = first
first.next = second
second.next = third
# third.next = fourth

ll.print()

print("Middle Element is :",ll.findMiddle())
```
